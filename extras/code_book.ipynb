```python
sns.set(style="whitegrid")  # Set the style of the plot

# Increase the chart width
plt.figure(figsize=(15, 6))

# Create the countplot
ax = sns.countplot(x='bathroom', data=train_df, hue='building_type', palette='viridis')

# Display the count on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points', rotation='vertical')  # Rotate the count labels vertically


# Customize the plot further
ax.set_title('Distribution of Bathroom Counts by Building Type')
ax.set_xlabel('Number of Bathrooms')
ax.set_ylabel('Count')

plt.show()
```

```python
import json

# Assuming train_df is your DataFrame
# Check for NaN values and replace them with an empty dictionary
train_df['amenities'] = train_df['amenities'].fillna('{}')
```

```python
# Convert the string representations to dictionaries
train_df['amenities'] = train_df['amenities'].apply(lambda x: json.loads(x.replace("'", "\"")))
```

```python
from geopy.geocoders import Nominatim

# Assuming train_df is your DataFrame and 'latitude' and 'longitude' are your columns
geolocator = Nominatim(user_agent="Location Finder")

# Define a function to get location information
def get_location_info(row):
    location = geolocator.reverse((row['latitude'], row['longitude']), language='en')
    if location:
        address = location.address.split(', ')
        return {
            'locality': address[-3] if len(address) >= 3 else None,
            'state': address[-2] if len(address) >= 2 else None
        }
    else:
        return {
            'locality': None,
            'state': None
        }


```
