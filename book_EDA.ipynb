```python
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

# Specify a custom user_agent and set the timeout
geolocator = Nominatim(user_agent="location_finder", timeout=20)

# Define a function to get location information
def get_location_info(row):
    try:
        location = geolocator.reverse((row['latitude'], row['longitude']), language='en').raw
        if location:
            return {
                'locality': location.get('address', {}).get('suburb', None),
                'city': location.get('address', {}).get('city', None)
            }
        else:
            return {
                'locality': None,
                'city': None
            }
    except GeocoderTimedOut:
        # Handle the timed-out error by returning default values
        return {
            'locality': None,
            'city': None
        }
```

# Outlier removal
- Property Size

```python
def remove_property_size_outliers(df):
    Q1 = df['property_size'].quantile(0.25)
    Q3 = df['property_size'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df[(df['property_size'] >= lower_bound) & (df['property_size'] <= upper_bound)]

    return df_filtered

# Apply the function to remove outliers
train_df_filtered = remove_property_size_outliers(train_df)
test_df_filtered = remove_property_size_outliers(test_df)
# Display the filtered DataFrame
print(train_df_filtered)
print(test_df_filtered)
```

```python
def remove_property_age_outliers(df):
    Q1 = df['property_age'].quantile(0.25)
    Q3 = df['property_age'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered_age = df[(df['property_age'] >= lower_bound) & (df['property_age'] <= upper_bound) & (df['property_age'] >= 0)]

    return df_filtered_age

# Apply the function to remove outliers and negative values
train_df_filtered_age = remove_property_age_outliers(train_df_filtered)
test_df_filtered_age = remove_property_age_outliers(test_df_filtered)

# Display the filtered DataFrame
print(train_df_filtered_age)
print(test_df_filtered_age)
```

# Outlier Removal on Balconies
- Removing balconies outlier on each category of building type

```python
def remove_balconies_outliers(df):
    # Group the DataFrame by building_type
    grouped = df.groupby('building_type')

    # Initialize an empty DataFrame to store the result
    result_df = pd.DataFrame()

    # Iterate over groups and remove outliers
    for group_name, group in grouped:
        Q1 = group['balconies'].quantile(0.25)
        Q3 = group['balconies'].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        filtered_group = group[(group['balconies'] >= lower_bound) & (group['balconies'] <= upper_bound)]

        # Append the filtered group to the result DataFrame
        result_df = pd.concat([result_df, filtered_group])

    return result_df

# Apply the function to remove outliers by building_type
train_df_filtered_balconies = remove_balconies_outliers(train_df_filtered_age)
test_df_filtered_balconies = remove_balconies_outliers(test_df_filtered_age)
# Display the filtered DataFrame
print(train_df_filtered_balconies)
print(test_df_filtered_balconies)
```

```python
train_df_filtered_balconies['cup_board'].value_counts()
```

```python
def remove_cupboard_outliers(df):
    # Initialize an empty DataFrame to store the result
    result_df = pd.DataFrame()

    for building_type, group in df.groupby('building_type'):
        Q1 = group['cup_board'].quantile(0.25)
        Q3 = group['cup_board'].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        filtered_group = group[(group['cup_board'] >= lower_bound) & (group['cup_board'] <= upper_bound)]

        # Concatenate the filtered group to the result DataFrame
        result_df = pd.concat([result_df, filtered_group])

    return result_df

# Apply the function to remove outliers by building_type
train_df_filtered_cupboard = remove_cupboard_outliers(train_df_filtered_balconies)
test_df_filtered_cupboard = remove_cupboard_outliers(test_df_filtered_balconies)

# Display the filtered DataFrame
print(train_df_filtered_cupboard)
print(test_df_filtered_cupboard)
```

```python
def remove_bathroom_outliers(df):
    result_df = pd.DataFrame()

    for building_type, group in df.groupby('type'):
        Q1 = group['bathroom'].quantile(0.25)
        Q3 = group['bathroom'].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        filtered_group = group[(group['bathroom'] >= lower_bound) & (group['bathroom'] <= upper_bound)]

        result_df = result_df.append(filtered_group)

    return result_df

# Apply the function to remove outliers by building_type

train_df_filtered_cupboard = remove_cupboard_outliers(train_df_filtered_cupboard)
test_df_filtered_cupboard = remove_cupboard_outliers(test_df_filtered_cupboard)
```
